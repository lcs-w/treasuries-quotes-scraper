```python
import requests
import json
import pandas as pd

from datetime import datetime
```

```python
def get_quotes_from_wsj(endpoint):
    headers = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}
    quotes = requests.get(endpoint, headers=headers)
    df = pd.DataFrame(quotes.json()["data"]["instruments"])
    df = df[
        [
            "maturityDate",
            "coupon",
            "bid",
            "ask",
            "change",
            "askYield",
        ]
    ]

    timestamp = quotes.json()["data"]["timestamp"]
    timestamp_dt = datetime.strptime(timestamp, "%A, %B %d, %Y")
    # timestamp_str = datetime.strftime(timestamp_dt, "%Y%m%d")

    return (df, timestamp_dt)
```

```python
endpoints = [
    "https://www.wsj.com/market-data/bonds/treasuries?id=%7B%22treasury%22%3A%22NOTES_AND_BONDS%22%7D&type=mdc_treasury",
    "https://www.wsj.com/market-data/bonds/treasuries?id=%7B%22treasury%22%3A%22BILLS%22%7D&type=mdc_treasury",
]

treasuries = get_quotes_from_wsj(endpoint=endpoints[0])
print(treasuries[1])
treasuries[0]
```
